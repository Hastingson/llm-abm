```python
import numpy as np
import matplotlib.pyplot as plt

N = 200
M = 200
np.random.seed(123)
group1 = np.random.normal(1, size=N)
group2 = np.random.normal(0, size=M)

def prototypicality(ind, groupa, groupb):
    I = groupa[ind]
    frac = (len(groupa)-1)/len(groupb)
    sum1 = np.sum(np.absolute(I - groupb))
    sum2 = np.sum(np.absolute(I - np.delete(groupa, [ind])))
    
    return frac * (sum1/sum2) 
    
```

```python
np.argmax([prototypicality(i, group2, group1) for i in range(M)])
```
